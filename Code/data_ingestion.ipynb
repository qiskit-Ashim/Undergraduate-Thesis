```python
import subprocess
import sys
import shutil
from pathlib import Path

# 1) Ensure Kaggle token location is set
# Expected file: /home/codespace/.kaggle/kaggle.json

# 2) Install Kaggle package in the active notebook environment
subprocess.run([sys.executable, "-m", "pip", "install", "-U", "kaggle"], check=True)

# 3) Resolve kaggle executable from the active environment
kaggle_exe = shutil.which("kaggle")
if kaggle_exe is None:
    kaggle_exe = str(Path(sys.executable).parent / "kaggle")

# 4) Download + unzip dataset (only if not already extracted)
project_dir = Path("/workspaces/Undergraduate-Thesis")
data_dir = project_dir / "artifacts"
data_dir.mkdir(parents=True, exist_ok=True)
dataset_root = data_dir / "BRAIN MRI"

if not dataset_root.exists():
    subprocess.run([
        kaggle_exe, "datasets", "download",
        "-d", "ashimdj89/brain-mri",
        "-p", str(data_dir),
        "--unzip"
    ], check=True)

# 5) Remove Windows/OS metadata files that are not images
metadata_names = {"desktop.ini", "thumbs.db", ".ds_store"}
removed = 0
for p in dataset_root.rglob("*"):
    if p.is_file() and p.name.lower() in metadata_names:
        p.unlink()
        removed += 1

print("Dataset root:", dataset_root)
print("Removed metadata files:", removed)
print("Class folders:", [p.name for p in dataset_root.iterdir() if p.is_dir()])
```
